## Ejercicio 2
Pide al usuario que ingrese un número. Si el número es **positivo** muestra 'Es positivo', si es **negativo** muestra 'Es negativo', si es **cero** muestra 'Es cero'.

Escribe tu código en la celda siguiente:

````python
numero_str = input("Por favor, ingresa un número: ")

try:
    numero = (numero_str)

    if numero > 0:
        print('Es positivo')
    elif numero < 0:
        print('Es negativo')
    else:
        print('Es cero')

except
    print("Entrada no válida. Por favor, ingresa un número.")
````
